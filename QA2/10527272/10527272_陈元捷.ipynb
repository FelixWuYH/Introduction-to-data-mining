```python
# PCA for dimensionality reduction 
#
from sklearn.decomposition import PCA
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
example_data = numpy.array([[0,1,2,3,4],[4,3,2,1,0]]) #
exa = numpy.array([[0,1,2,3,4],[4,3,2,1,0]]) #

#step 

# 1.
#2.
#3.
#4.k
#5.

avg = numpy.mean(example_data,0) # 
plt.plot(example_data)
plt.title("First data before PCA")
plt.show()

i = 0
j = 0
while i < 2:        #
    j = 0
    while j < 5:
        example_data[i,j] = example_data[i,j] - avg[j]
        j = j + 1
    i = i + 1

ex_cov = numpy.cov(example_data)               #

        
e_value,e_vector = numpy.linalg.eig(ex_cov) #,



sort_e_value = numpy.argsort(-e_value)
cond = e_vector[sort_e_value[1]]  # 

final_Data = numpy.dot(example_data.T,cond)# 
#print(final_Data) #


pca = PCA(n_components=1)  # 

exa = pca.fit_transform(exa.T)
plt.plot(final_Data)
plt.title("First data after PCA")
plt.show()
#print(exa)

#
exb = numpy.array([[2.5,0.5,2.2,1.9,3.1,2.3,2,1,1.5,1.1],[2.4,0.7,2.9,2.2,3.0,2.7,1.6,1.1,1.6,0.9]])
exb_2 = numpy.array([[2.5,0.5,2.2,1.9,3.1,2.3,2,1,1.5,1.1],[2.4,0.7,2.9,2.2,3.0,2.7,1.6,1.1,1.6,0.9]])
avg_b = numpy.mean(exb,axis=1) # 
plt.plot(exb)
plt.title("Second data before PCA")
plt.show()
i = 0
j = 0
while i < 2:        #
    j = 0
    while j < 10:
        exb[i,j] = exb[i,j] - avg_b[i]
        j = j + 1
    i = i + 1


exb_cov = numpy.cov(exb)               #
exb_value,exb_vector = numpy.linalg.eig(exb_cov) #,


cond_b = numpy.transpose(exb_vector)[1]  # 



final_Datab = numpy.dot(exb.T,cond_b)
plt.plot(final_Datab)
plt.title("Second data after PCA")
plt.show()
#print(final_Datab)

pcb = PCA(n_components=1)  # 

exb1 = pcb.fit_transform(exb_2.T)

#print(exb1)
```
