```python
# PCA for dimensionality reduction 主成分分析降维
#目标：最大限度保留数据特征的同时降低维度
from sklearn.decomposition import PCA
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
example_data = numpy.array([[0,1,2,3,4],[4,3,2,1,0]]) #书上范例资料，目标：二维→一维
exa = numpy.array([[0,1,2,3,4],[4,3,2,1,0]]) #书上范例资料，目标：二维→一维

#step 

# 1.对数据中心化
#2.求特征的协方差矩阵
#3.求协方差矩阵的特征值和特征向量
#4.取最大的k个特征值所对应的特征向量
#5.将样本点投影到选取的特征向量上

avg = numpy.mean(example_data,0) # 取每一维度平均值
plt.plot(example_data)
plt.title("First data before PCA")
plt.show()

i = 0
j = 0
while i < 2:        #对每一元素做中心化
    j = 0
    while j < 5:
        example_data[i,j] = example_data[i,j] - avg[j]
        j = j + 1
    i = i + 1

ex_cov = numpy.cov(example_data)               #求协方差矩阵

        
e_value,e_vector = numpy.linalg.eig(ex_cov) #协方差矩阵特征值,特征向量



sort_e_value = numpy.argsort(-e_value)
cond = e_vector[sort_e_value[1]]  # 选取较大的特征值对应的特征向量

final_Data = numpy.dot(example_data.T,cond)# 计算投影结果
#print(final_Data) #输出结果


pca = PCA(n_components=1)  # 验证结果

exa = pca.fit_transform(exa.T)
plt.plot(final_Data)
plt.title("First data after PCA")
plt.show()
#print(exa)

#其他数据
exb = numpy.array([[2.5,0.5,2.2,1.9,3.1,2.3,2,1,1.5,1.1],[2.4,0.7,2.9,2.2,3.0,2.7,1.6,1.1,1.6,0.9]])
exb_2 = numpy.array([[2.5,0.5,2.2,1.9,3.1,2.3,2,1,1.5,1.1],[2.4,0.7,2.9,2.2,3.0,2.7,1.6,1.1,1.6,0.9]])
avg_b = numpy.mean(exb,axis=1) # 取每一维度平均值
plt.plot(exb)
plt.title("Second data before PCA")
plt.show()
i = 0
j = 0
while i < 2:        #对每一元素做中心化
    j = 0
    while j < 10:
        exb[i,j] = exb[i,j] - avg_b[i]
        j = j + 1
    i = i + 1


exb_cov = numpy.cov(exb)               #求协方差矩阵
exb_value,exb_vector = numpy.linalg.eig(exb_cov) #协方差矩阵特征值,特征向量


cond_b = numpy.transpose(exb_vector)[1]  # 选取较大的特征值对应的特征向量



final_Datab = numpy.dot(exb.T,cond_b)
plt.plot(final_Datab)
plt.title("Second data after PCA")
plt.show()
#print(final_Datab)

pcb = PCA(n_components=1)  # 验证结果

exb1 = pcb.fit_transform(exb_2.T)

#print(exb1)
```
